<div align="center"> <h1>Model the system</h1></div>

$$\cfrac{dR_1(x)}{dx} + \delta \cdot R_1(x) = R_2(x) \cdot R_3(x)$$

$$\cfrac{dR_2(x)}{dx} + \delta \cdot R_2(x) = R_1(x) \cdot R_3(x)$$

$$\cfrac{dR_3(x)}{dx} + \delta \cdot R_3(x) = - R_1(x) \cdot R_2(x)$$

<div align="center"> <b> with conditions:</b>
$\delta = 0.05; R_1(0) = R_2(0) = 0.05; R_3(0) = 1; x \in [0,100].$</div>

```python
# imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.integrate import odeint

sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 12})
```

```python
# system of differential equations
def model(r, t):
    x = r[0]
    y = r[1]
    z = r[2]
    
    dxdt = y * z - delta * x
    dydt = x * z - delta * y
    dzdt = -x * y - delta * z
    drdt = [dxdt, dydt, dzdt]
    return drdt
```

```python
# initial conditions
r0 = [0.05, 0.05, 1]

# params
delta = 0.05
t = np.linspace(0, 100, 1000)

# sloving system of differential equations
r = odeint(model, r0, t)
```

```python
# plotting graphs
plt.figure(figsize = (16, 8))
plt.plot(t, r[:,0], color = '#4d83f6', lw = 2.3, label = '$R_1$')
plt.plot(t, r[:,1], color = '#ec5e5e', lw = 2.3, label = '$R_2$')
plt.plot(t, r[:,2], color = '#16e1c7', lw = 2.3, label = '$R_3$')
plt.legend()
```
