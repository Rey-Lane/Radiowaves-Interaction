```python
# imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import integrate

sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 12})
```

```python
# params
delta, tau = 0, 1
gammas = [0.1, 0.22, 0.6]
time = np.arange(-5, 10, 0.05)
values = []
```

```python
def A(n):
    return (complex(n + delta, gamma) / complex(n + delta, -gamma)) * np.exp(complex(-n * n / 2, n * (t / tau)))

# calculate
for gamma in gammas:
    for t in time:
       values.append(abs(integrate.nquad(A, [[-5, 5]])[0]))
```

```python
# plotting graphs
plt.figure(figsize = (16, 8))
plt.plot(time, values[:len(time)], label = '$\gamma = 0.1$', color = '#4d83f6', lw = 2.3)
plt.plot(time, values[len(time):2*len(time)], label = '$\gamma = 0.22$', color = '#ec5e5e', lw = 2.3)
plt.plot(time, values[2*len(time):3*len(time)], label = '$\gamma = 0.6$', color = '#16e1c7', lw = 2.3)
plt.legend()
```
