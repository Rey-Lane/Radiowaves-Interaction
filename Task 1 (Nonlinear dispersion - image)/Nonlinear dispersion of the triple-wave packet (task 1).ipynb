```python
# imports
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 14})
```

```python
# params
graphs = []
Ao, F = 1, 0.1
z = 1/np.sqrt(2)
beta = np.radians(np.linspace(0, 270, 270))
```

```python
# functions (10, 11, 12 equations in theory article)
def Pn(m, An, Ab):
    return -z**2/8 * (An * (An**2 * (1 - F)*(1 + 2*F) + 2 * Ab**2 * (1 - 3*F - 2 * F**2) + 2 * Ao**2 * (1 - F) ) * np.sin(np.pi/2) + Ab*Ao**2 * (1-3*F) * np.sin(np.pi/2 - 2 * beta))
    
def Pb(m, An, Ab):
    return -z**2/8 * (Ab * (Ab**2 * (1 + F)*(1 - 2*F) + 2 * An**2 * (1 + 3*F - 2 * F**2) + 2 * Ao**2 * (1 + F) ) * np.sin(np.pi/2) + An*Ao**2 * (1+3*F) * np.sin(np.pi/2 - 2 * beta))

def Po(m, An, Ab):
    return -z**2/8 * Ao * ((An**2 * (3 -(2 * F - 1)**2) + Ab**2 * (3 -(2 * F + 1)**2) + Ao**2) * np.sin(np.pi/2) + 2 * An*Ab * (1 - 2 * F**2) * np.sin(np.pi/2 + 2 * beta))    
```

```python
# calculate functions for different params

# m = 0.5; An=Ab=0.25;
Pn_025 = Pn(0.5, 0.25, 0.25)
Pb_025 = Pb(0.5, 0.25, 0.25)
Po_025 = Po(0.5, 0.25, 0.25)

# m=2; An=Ab=1;
Pn_1 = Pn(2, 1, 1)
Pb_1 = Pb(2, 1, 1)
Po_1 = Po(2, 1, 1)
```

```python
# normalization
graphs.append(Pn_025/Pn_025[0])
graphs.append(Pb_025/Pb_025[0])
graphs.append(Po_025/Po_025[0])

graphs.append(Pn_1/Pn_1[0])
graphs.append(Pb_1/Pb_1[0])
graphs.append(Po_1/Po_1[0])
```

```python
# plotting graphs
plt.figure(figsize = (18, 6))
for i in graphs:
    plt.plot(i, lw = 2)

plt.xlabel('$\\beta_0$, deg')
plt.legend(['Pn: m = 0.5', 'Pb: m = 0.5', 'Po: m = 0.5', 'Pn: m = 2', 'Pb: m = 2', 'Po: m = 2'])
```
